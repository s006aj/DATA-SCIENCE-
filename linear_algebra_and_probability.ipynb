IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

LOADING DATASET

```python
data_url = 'https://drive.google.com/uc?export=download&id=1OnJ8qjX8HAlNU5JCHFRdYjMHLd78Ft56'
df=pd.read_csv(data_url, delimiter=';')
df.head()
```

HANDLING MISSING DATA

```python
df.fillna(df.mean(),inplace = True)
print('data loaded')
print(f'shape of data :{df.shape}')
df.head()
```

EXTRACTING COLOUMNS ALCOHOL AND CITRIC ACID AS VECTOR



```python
alcohol_vector=df['alcohol'].to_numpy()
citric_acid_vector=df['citric acid'].to_numpy()
```

COVARIENCE MATRIX

```python
X=df[['alcohol','density']].to_numpy()
covarience_matrix=np.cov(X.T)
print('covarience_matrix')
print(covarience_matrix)
```

EIGEN DECOMPOSITION

```python
eigenvalues,eigenvectors =np.linalg.eig(covarience_matrix)
print('eigen values')
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
```

 MOST COMMON VINE QUALITY

```python
most_common_quality=df['quality'].mode()[0]
quality_count= df['quality'].value_counts().sort_index()
print('most_common_wine_quality')
print(most_common_quality)
print('wine quality count')
print(quality_count)

```

```python
df['quality'].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Wine Quality Scores')
plt.xlabel('Quality Score')
plt.ylabel('Count')
plt.show()
```

from the graph,we can see that wine quality with 5 and 6 has most number of wine.Wines with lesser quality have lesser number.
